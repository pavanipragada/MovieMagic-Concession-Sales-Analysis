### Answer 11.i [1 point] Write code to compute and add a new column in popDF that contains the compound annual growth rate of population between 2014 and 2019. Call this new column ‘2019_5cagr’. 

```python
n = 2019-2014
popDF['2019_5cagr'] = np.round(((popDF['y2019']/popDF['y2014'])**(1/n))-1,4)
popDF['2019_5cagr'].head()
```

```python
# 11b. More complex operations on data for which built-in functions may not exist – using DF.apply()
popDF[['y2019']].apply(lambda x: np.exp((x - x.mean())/x.std())).head()
```

```python
def expstandardize(x):
    return np.exp((x-x.mean())/x.std())
popDF[['y2019']].apply(expstandardize).head()
```

```python
popDF[['y2018','y2019']].apply(expstandardize).head()
```

### Answer 11.ii. [2 points] Write code to compute a new column in popDF that contains the coefficient of variation for each state for the population in years 2010-2019.

```python
popDF['coeffofvar'] = popDF.loc[:,'y2010':'y2019'].apply(lambda x: (x.std()/x.mean()) , axis =1)
popDF['coeffofvar'].head()
```

```python
# 11c. Aggregation operations using DF.agg()
popDF[['y2018','y2019']].agg(['count','sum','max','min','mean','var','skew','kurt','median','idxmax','idxmin'])
```

```python
popDF[['y2018','y2019']].agg({'y2019':['max','min'],'y2018':['var','std']})
```

```python
popDF[['y2018','y2019']].agg({'y2019':['max','min'],'y2018':['var','std',np.min,np.max]})
```

```python
# 11d. Performing grouping operations using DF.groupby() 
popDF[['Region','y2010','y2019']].groupby(['Region']).mean().sort_values(['Region'])
```

```python
popDF.groupby(['Region'])[['y2019','y2010']].agg(['min','idxmin','max','idxmax','mean'])
```

### Answer 11.iii. [2 points] Write code to compute a new column in popDF that contains the mean of the population in year 2019 grouped by region (Midwest, Northeast, South, West). Call this new column ‘mean2019popbyReg’. 

```python
popDF['mean2019popbyReg'] = popDF.groupby(['Region'])['y2019'].transform('mean')
```
